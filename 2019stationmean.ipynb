```python
import findspark
findspark.init()
from pyspark import SparkContext
sc=SparkContext()
sc
```

```python
from pyspark.sql import SparkSession
spark=SparkSession.builder.getOrCreate()
spark
```

```python
data2019 = spark.read.option("header", "true").csv("s3://open-data-noaa/cleaneddata/*.csv")
```

```python
data2019.cache()
```

```python
data2019.printSchema()
```

```python
data2019.show(5)
```

```python
data2019.describe().toPandas().transpose()
```

```python
data2019.groupBy("station").agg({'latitude':'avg', 'longitude':'avg','elevation':'avg', 'wind':'avg', 'cig':'avg', 'vis':'avg', 'temp':'avg', 'dew':'avg', 'slp':'avg', 'liquid':'avg'}).show()
```

```python
data=data2019.groupBy("station").agg({'latitude':'avg', 'longitude':'avg','elevation':'avg', 'wind':'avg', 'cig':'avg', 'vis':'avg', 'temp':'avg', 'dew':'avg', 'slp':'avg', 'liquid':'avg'})
```

```python
data.describe().toPandas().transpose()
```

```python
data.show()
```

```python
data.write.csv("s3://open-data-noaa/2019data_mean", sep=',', header=True,mode="overwrite")
```
